```python
# quick & dirty way of importing from parent directory
import sys
sys.path.append('..')

import detect_books_from_shelf
import image_title_author_scrape
import google_books_api

import cv2
```

```python
def get_rect_from_img(img_path):
    img = cv2.imread(img_path)
    booklines = detect_books_from_shelf.get_book_lines(img, debug=False)
    rectangles = detect_books_from_shelf.find_rectangles(booklines)
    print("%d rectangles identified" % len(rectangles))
    return rectangles, img


def get_text_from_rect(rectangles, img):
    books_img = []
    for x in rectangles:
        books_img.append(detect_books_from_shelf.make_crops_from_rect(img, x))

    storage_book = {}
    for order, book in enumerate(books_img):

        filename = "{}_ind.png".format(order)
        cv2.imwrite(filename, book)
        text = image_title_author_scrape.detect_text(filename)
        if text:
            corrected_text = text.split('\n')
            #corrected_text = [correct_text(x) for x in corrected_text]
            corrected_text.append(filename)
            storage_book.setdefault(order, corrected_text)

    return storage_book
```

```python
# identify the spines of the books
#rectangles, img = get_rect_from_img("prova.jpg")
#rectangles, img = get_rect_from_img("/tmp/index.jpeg")
rectangles, img = get_rect_from_img("tanti_libri.jpg")
# look for the text on each spine
storage_book = get_text_from_rect(rectangles, img)
print("detected texts from books are the following {}".format(storage_book))
```
